```python
path = '/data/AlgProj/tct_yaoms/model/resnet18_0716_error_analysis.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
csv = pd.read_csv(path)
```

```python
csv.head()
```

```python
# figure1, ax1 = plt.subplots(2,5)
fig=plt.figure(figsize=(12, 12))

for i in range(10):
    img_path = csv.loc[i, 'img_path']
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig.add_subplot(2, 5, i+1)
    plt.imshow(img)
plt.show()
```

```python
fig=plt.figure(figsize=(24, 16))

for i in range(20):
    img_path = csv.loc[i+10, 'img_path']
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig.add_subplot(4, 5, i+1)
    plt.imshow(img)
plt.show()
```
