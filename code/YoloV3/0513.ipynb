```python
import pandas as pd
import json
```

```python
csv_path = '/data/AlgProj/niezl/TCT/TCT_Data/Detection/0513/tct_20190513.csv'
```

```python
csv = pd.read_csv(csv_path)
csv.tail()
```

```python
label_dir = '/data/AlgProj/tct_yaoms/PyTorch-YOLOv3/data/custom/labels/'
img_dir = '/data/AlgProj/tct_yaoms/PyTorch-YOLOv3/data/custom/images/'
df = pd.DataFrame(columns=['img_path'])
shape = (1936, 1216)
for idx in csv.index:
    img_name = csv.loc[idx, 'seriesID']
    save_name = img_name.split('.')[0] + '.txt'
    x_min = csv.loc[idx, 'xmin']
    y_min = csv.loc[idx, 'ymin']
    x_max = csv.loc[idx, 'xmax']
    y_max = csv.loc[idx, 'ymax']
    
    img_path = img_dir + img_name
    df = df.append([{'img_path':img_path}], ignore_index=True)
    
    center_x = ((x_min + x_max)/2.0)/shape[0]
    center_y = ((y_min + y_max)/2.0)/shape[1]
    width = (x_max - x_min)/shape[0]
    heith = (y_max - y_min)/shape[1]
    
    save_path = label_dir + save_name
    string =  '0' + ' ' + str(center_x) + ' ' + str(center_y) + ' ' + str(width) + ' '+ str(heith)
    with open(save_path, 'a') as f:
        f.write(string)
        f.write('\n')
```
