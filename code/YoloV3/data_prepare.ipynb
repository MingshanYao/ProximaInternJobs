```python
csv_path = '/data/AlgProj/tct_yaoms/data/tct_0513/detail_infomation.csv'
data_dir = '/data/AlgProj/niezl/TCT/1_DataBase/20190711/'
```

```python
csv = pd.read_csv(csv_path)
```

```python
len(csv['影像结果编号'].unique()) == len(csv)
```

```python
csv.tail()
```

```python
df = pd.DataFrame(columns=['img_path'])
for idx in csv.index:
    series = csv.loc[idx, '序列编号']
    result_idx = csv.loc[idx, '影像结果编号']
    z = str(csv.loc[idx,'z'])
    len_z = len(z)
    img_name = '0'*(6-len_z) + z + '.jpg'
    img_path = data_dir + series + '/' + img_name
#     x_min = csv.loc[idx, 'x_min']
#     y_min = csv.loc[idx,'y_min']
#     x_max = csv.loc[idx,'x_max']
#     y_max = csv.loc[idx, 'y_max']
    
    img = Image.open(img_path)
    shape = img.size
    
    x_min = max(0, csv.loc[idx, 'x_min'])
    y_min = max(0, csv.loc[idx, 'y_min'])
    x_max = min(shape[0]-1, csv.loc[idx, 'x_max'])
    y_max = min(shape[1]-1, csv.loc[idx, 'y_max'])
    center_x = ((x_min + x_max)/2.0)/shape[0]
    center_y = ((y_min + y_max)/2.0)/shape[1]
    width = (x_max - x_min)/shape[0]
    heith = (y_max - y_min)/shape[1]
    
    img_save_path = '/data/AlgProj/tct_yaoms/PyTorch-YOLOv3/data/custom/images/'+str(result_idx)+'.jpg'
    labels_save_path = '/data/AlgProj/tct_yaoms/PyTorch-YOLOv3/data/custom/labels/'+str(result_idx)+'.txt'
    img.save(img_save_path, quality=95)
    df = df.append([{'img_path':img_save_path}], ignore_index=True)
    
    write_string = '0' + ' ' + str(center_x) + ' ' + str(center_y) + ' ' + str(width) + ' '+ str(heith)
    with open(labels_save_path, 'w') as f:
        f.write(write_string)
```
