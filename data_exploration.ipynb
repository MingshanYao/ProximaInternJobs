```python
import SimpleITK as sitk
import cv2
import matplotlib.pyplot as plt
import glob
import sys
from tqdm import tqdm
import os
import nibabel as nib
from nibabel.viewers import OrthoSlicer3D

data_parent = "/data/brain_7_cta/"
data_sib_lst = glob.glob(data_parent+'*')
print(len(data_sib_lst))
```

```python
def nib_2_jpg(index):
    data_paths = glob.glob(data_sib_lst[index]+'/*.nii.gz')
    for path in data_paths:
        print(path)
        cur_dir = path.split('/')[-2]
        cur_name = path.split('/')[-1].split('.')[-3]
        cur_type = cur_name.split('_')[3]
        
        target_path  = os.path.join(data_parent,cur_dir,cur_type)
        #print(target_path)
        if not os.path.exists(target_path):
            os.mkdir(target_path)
        img = nib.load(path)    
        #print(cur_name,cur_type)
        #print(img)
        #print(img.header['db_name'])    #输出nii的头文件
        width, height, queue = img.dataobj.shape
        #print(width, height, queue)
        #nib.viewers.OrthoSlice3D(img.dataobj).show()
        num = 0
        for i in range(0, queue, 1):
            img_arr = img.dataobj[:,:,i]
            #plt.subplot(5,4,num)
            #plt.imshow(img_arr, cmap='gray')
            save_name = os.path.join(data_parent,cur_dir,cur_type,str(num)+'.jpg')
            print(save_name)
            cv2.imwrite(save_name,img_arr)
            num += 1
        #plt.show()
    #print("done!")
```

```python
for i in range(len(data_sib_lst)):
    nib_2_jpg(i)
```
